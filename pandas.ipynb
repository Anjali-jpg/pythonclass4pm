```python
#DictReader #dictWriter
```

```python
import csv                                        #with open() doesn't need close of file
with open("data.csv") as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        print(row)
```

```python
import csv                                #changed row of list to dict         
with open("data.csv") as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        print(dict(row))
```

```python
import csv                                     #dict to list by appending
data=[]
with open("data.csv") as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        data.append (dict(row))
data
```

```python
import csv                                     #dict to list by appending
data=[]
b=open("data.csv")
reader = csv.DictReader(b)
for row in reader:
    data.append (dict(row))
data
```

```python
import csv
csv_column=['Name','Age','Address']
dict=[
    {"Name":"ram","Age":77,"Address":"Kathmandu"},
    {"Name":"shyam","Age":97,"Address":"patan"}
]
b=open("data.csv",'w')
writer=csv.DictWriter(b, fieldnames=csv_column)
writer.writeheader()
for data in dict:
    writer.writerow(data)
b.close()  
```

```python
import pandas as pd
d = {'Name': ['ram', 'shyam'], 'Ntc': [768756465465, 877667575], 'Ncell': [7676665454, 67655445]}
d1={'Name': ['ram', 'shyam'], 'Ntc': [768756465465, 877667575], 'Ncell': [7676665454, 67655445]}
d2={'Name': ['ram', 'shyam'], 'Ntc': [768756465465, 877667575], 'Ncell': [7676665454, 67655445]}

df1=pd.DataFrame(d)
df2=pd.DataFrame(d1)
df3=pd.DataFrame(d2)
df=pd.concat([df1,df2,df3])

df
```
