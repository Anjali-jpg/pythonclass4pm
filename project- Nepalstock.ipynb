```python
from bs4 import BeautifulSoup
import requests
url = BeautifulSoup('http://www.nepalstock.com/todaysprice','html.parser')
response = requests.get(url)
response
```

# New Section

```python
data = response.text
data
```

```python
soup = BeautifulSoup(data,'lxml')
soup
```

```python
soup_table = soup.table
soup_table
```

```python
tag = soup_table.find_all('tr')
tag
```

```python
a = []

for i in tag:
  x = i.get_text()
  a.append(x)

print(a)
```

```python
a
```

```python
y = []
for i in a:
  x = i.split('\n')[1:-1]
  y.append(x)

print(y)
```

```python
import csv
f = open('Nepalstock.csv','w')
write = csv.writer(f)
for i in y:
  write.writerow(i)

f.close()
```
